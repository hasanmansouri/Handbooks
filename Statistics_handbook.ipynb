#  2. Mean, Median, and Mode



### Definitions

- **Mean (Average)**: Add all the values and divide by how many there are. 
- **Median**: The middle value in sorted data. If there are 2 middle values, we take their average.
- **Mode**: The value that appears most often in the dataset.



### When to Use Each?

- Use **mean** when the data is not skewed (balanced).
- Use **median** when the data has outliers or is skewed.
- Use **mode** when working with categories or repeated values.



### Formulas

- **Mean** = (x₁ + x₂ + ... + xₙ) / n  
- **Median** = Middle value after sorting (or average of two middles if even number of items)  
- **Mode** = Most frequent value




And here we see how to use each in the code with an example (Python)

```python
import numpy as np
from scipy import stats

# Example dataset
data = np.array([4, 5, 6, 8, 10, 10, 12, 12, 12, 14])

# Mean
mean_val = np.mean(data)
print("Mean:", mean_val)

# Median
median_val = np.median(data)
print("Median:", median_val)

# Mode
mode_val = stats.mode(data, keepdims=True)
print("Mode:", mode_val.mode[0], " (appears", mode_val.count[0], "times)")
```

### Visualizing the Distribution
We can use a **histogram** to visualize the frequency of values.


```python
plt.figure(figsize=(8,4))
sns.histplot(data, bins=8, kde=True, color="skyblue")
plt.title("Histogram of Data Values")
plt.xlabel("Values")
plt.ylabel("Frequency")
plt.axvline(mean_val, color='green', linestyle='--', label=f"Mean: {mean_val}")
plt.axvline(median_val, color='orange', linestyle='--', label=f"Median: {median_val}")
plt.axvline(mode_val.mode[0], color='red', linestyle='--', label=f"Mode: {mode_val.mode[0]}")
plt.legend()
plt.show()
```

# 3. Range, Variance, and Standard Deviation (VERY IMPORTANT)

---

### Definitions

---



### **Range**: The difference between the highest and lowest values.  
  > Tells us how spread out the data is in general.

Formula: **Range** = max(x) – min(x)


---


### **Variance**: The average of the squared differences from the Mean.  
  > It measures how far each number is from the average.

Formula:
**Variance (Population)** = Σ(x - μ)² / N (Don't worry It's much simpler in the code!)

**Variance (Sample)** = Σ(x - x̄)² / (n - 1) (Don't worry It's much simpler in the code!)


---


### **Standard Deviation (SD)**: The square root of variance.  
  > It tells us how much the values typically differ from the mean.

Formula:
**Standard Deviation** = √Variance

---


### Why Do These Matter?

These values show how **spread out** or **concentrated** your data is.

- **Low Standard Deviation**: Data is close to the mean (less variation).
- **High Standard Deviation**: Data is more spread out (more variation).




---

> In practice, we mostly use NumPy for these calculations. **No need to manually use formulas**.


```python
# Sample dataset
data = np.array([4, 5, 6, 8, 10, 10, 12, 12, 12, 14])

# Range
range_val = np.max(data) - np.min(data)
print("Range:", range_val)



# Population variance & std
var_pop = np.var(data)
std_pop = np.std(data)

# Sample variance & std
var_sample = np.var(data, ddof=1)
std_sample = np.std(data, ddof=1)



print("Population Variance:", var_pop)
print("Population Std Dev:", std_pop)
print("Sample Variance:", var_sample)
print("Sample Std Dev:", std_sample)
```

### Visualizing Spread with a Boxplot

A boxplot helps us see the **range**, **median**, and **outliers**.


```python
plt.figure(figsize=(6, 4))
sns.boxplot(x=data, color="lightblue")
plt.title("Boxplot of Data")
plt.xlabel("Value")
plt.show()
```

### What Does This Boxplot Show?

A **boxplot** is a great way to see how the data is spread out. Here's what it means:

- The **box** shows the middle 50% of the data (called the Interquartile Range, or IQR).
- The **line inside the box** is the **median** (middle value).
- The **left edge** of the box is the **25th percentile** (Q1).
- The **right edge** is the **75th percentile** (Q3).
- The **"whiskers"** are the lines that stretch to the smallest and largest values *within a normal range*.
- Any **dots outside the whiskers** are **outliers** — values that are unusually far from the rest.

---

In short:  
It’s a compact way to see:
- If your data is **centered** or **skewed**
- If there are any **outliers**
- How **spread out** the values are


# 4. Probability

---

### What is Probability?

**Probability** tells us how likely something is to happen.  
It’s a number between **0 and 1**, where:

- 0 means **impossible**
- 1 means **certain**
- Values in between show how likely the event is


---


###  Formula for Basic Probability:


$P(A) = \frac{\text{Number of favorable outcomes}}{\text{Total number of outcomes}}$


 *Don't worry — in the code it’s much simpler! (I wanna say it everytime)*


---


### Addition Rule (For "A or B"):

If A and B are two events:


$P(A \cup B) = P(A) + P(B) - P(A \cap B)$


This avoids double-counting the overlap.

 *Don't worry — in the code it’s much simpler!*


---


### Multiplication Rule (For "A and B"):

#### If A and B are **independent**:


$P(A \cap B) = P(A) \times P(B)$


#### If A and B are **dependent**:


$P(A \cap B) = P(A) \times P(B|A)$


 *Don’t worry — we’ll show it step by step with examples!*


---


### Why do we Learn Probability?

Probability is **everywhere**:
- Weather forecasts (chance of rain )
- Machine learning (predicting outcomes)
- Games (rolling dice)
- Risk analysis in finance or health

Understanding it helps you **make better decisions** with data!


```python
#Just to see
# Example: What is the probability of getting a 3 when rolling a 6-sided die?

favorable = 1  # Only one "3"
total = 6      # 6 possible outcomes

prob = favorable / total
print("P(rolling a 3):", prob)


# Example: What is the probability of drawing a red card or an Ace from a standard 52-card deck?


# 26 red cards (hearts + diamonds), 4 Aces, but 2 are red aces
P_red = 26 / 52
P_ace = 4 / 52
P_red_ace = 2 / 52

P_red_or_ace = P_red + P_ace - P_red_ace
print("P(Red or Ace):", P_red_or_ace)


# Example: Tossing two fair coins — what's the probability of getting two heads?

P_head = 0.5
P_two_heads = P_head * P_head
print("P(Two Heads):", P_two_heads)
```

# 5. Distributions


---


### What is a Distribution?

A **distribution** shows how data is spread out — how often values occur. (We've already talked about it)

In data science, we often work with these:


---


## 1. Normal Distribution

Also called the **bell curve** because of its shape.

- Symmetrical
- Most data near the center
- Mean = Median = Mode

This is very common in real life:
- Heights of people
- Test scores
- Measurement errors


---


### Formula:

The formula is: $f(x) = \frac{1}{\sqrt{2\pi\sigma^2}} e^{ -\frac{(x - \mu)^2}{2\sigma^2} }$


- \( \mu \) = mean
- \( \sigma \) = standard deviation

 *Don’t worry, in the code it's much simpler!*


---


## 2. Binomial Distribution

- For things with **2 outcomes**: success/failure, yes/no, 1/0  
- You repeat the experiment a fixed number of times

Examples:
- Tossing a coin 10 times
- Checking how many people say "yes" in a survey


---


### Formula:


$P(X = k) = \binom{n}{k} p^k (1 - p)^{n - k}$


Where:
- \( n \) = total trials
- \( k \) = number of successes
- \( p \) = probability of success

 *Don’t worry — Python has built-in tools to do this easily!*


---


## 3. Histograms

- Used to **visualize distributions**
- Shows **how often** values fall into **intervals**
- Great for checking if data looks normal

We'll see this with real plots next:


```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import binom, norm

# Set style
sns.set(style='whitegrid')

# Normal Distribution
mean = 0
std_dev = 1
x = np.linspace(-4, 4, 1000)
y = norm.pdf(x, mean, std_dev)

plt.figure(figsize=(8, 4))
plt.plot(x, y, color='blue')
plt.title(" Normal Distribution (mean=0, std=1)")
plt.xlabel("Value")
plt.ylabel("Probability Density")
plt.grid(True)
plt.show()

# Binomial Distribution
n = 10  # trials
p = 0.5  # success probability
x = np.arange(0, n+1)
y = binom.pmf(x, n, p)

plt.figure(figsize=(8, 4))
sns.barplot(x=x, y=y, color='green')
plt.title(" Binomial Distribution (n=10, p=0.5)")
plt.xlabel("Number of Successes")
plt.ylabel("Probability")
plt.grid(True)
plt.show()

# Histogram of random data
data = np.random.normal(loc=0, scale=1, size=1000)

plt.figure(figsize=(8, 4))
plt.hist(data, bins=30, color='orange', edgecolor='black')
plt.title(" Histogram of Normally Distributed Data")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.grid(True)
plt.show()
```

### What Did We Just See?

- **Normal Distribution Plot**: The smooth bell curve shows how values are centered around the mean. Most values are near 0.

- **Binomial Distribution**: We simulated tossing a coin 10 times and seeing how likely it is to get 0 to 10 heads.

- **Histogram**: We generated 1000 values from a normal distribution and showed how often each value range appears.

These charts help you **understand your data** before building models. That’s why every data scientist needs to know distributions!


##  Correlation and Causation

**Correlation** is a statistical measure that describes the extent to which two variables change together.

- If two variables tend to **increase or decrease together**, they have a **positive correlation**.
- If one increases while the other decreases, they have a **negative correlation**.
- If there's **no clear pattern**, the correlation is weak or nonexistent.

 **Causation** means one variable *actually causes* a change in the other — but just because two things are correlated doesn't mean one causes the other!

>  **"Correlation does not imply causation."**

---
##  Pearson Correlation Coefficient Formula


$r = \frac{ \sum (x_i - \bar{x})(y_i - \bar{y}) }{ \sqrt{ \sum (x_i - \bar{x})^2 } \cdot \sqrt{ \sum (y_i - \bar{y})^2 } }$


Where:
- \( r \) = Pearson correlation coefficient  
- \( x_i, y_i \) = individual values  
- \( \bar{x}, \bar{y} \) = means of x and y

 *Don’t worry, in the code it’s much simpler!*

---
##  Code Example: Calculating Correlation with NumPy & Pandas


```python
#import numpy as np
#import pandas as pd
#import matplotlib.pyplot as plt

# Sample data
study_hours = [1, 2, 3, 4, 5, 6, 7]
exam_scores = [50, 55, 65, 70, 75, 80, 88]

# Correlation using numpy
correlation = np.corrcoef(study_hours, exam_scores)[0, 1]
print("Pearson Correlation Coefficient:", correlation)

# Using pandas
df = pd.DataFrame({'Study Hours': study_hours, 'Exam Scores': exam_scores})
print("\nPandas Correlation Matrix:")
print(df.corr())
```

Let's visualize the relationship using a scatter plot.


```python
plt.scatter(study_hours, exam_scores, color='teal')
plt.title("Study Hours vs Exam Scores")
plt.xlabel("Study Hours")
plt.ylabel("Exam Scores")
plt.grid(True)
plt.show()
```

## Explanation:

In the plot above, you can see that as study hours increase, exam scores also increase. That’s a **positive correlation**.

The correlation coefficient (close to 1) confirms a **strong positive relationship**.

But remember:  **this doesn't prove studying more *causes* better scores!** There could be other factors involved.

##  Linear Regression

**Regression** is used to model the relationship between a **dependent variable** (what you want to predict) and one or more **independent variables** (the inputs).

In **simple linear regression**, we use one input to predict an output with a **straight line**.

---

##  Linear Regression Formula


$y = mx + b$


Where:
- \( y \) = predicted value (dependent variable)
- \( x \) = input (independent variable)
- \( m \) = slope of the line
- \( b \) = y-intercept

 *Don’t worry, in the code it’s much simpler!*

---

##  What does it mean?

- The **slope** (m) tells us how much y changes for each unit change in x.
- The **intercept** (b) is the value of y when x = 0.


```python
import numpy as np
import matplotlib.pyplot as plt

# Example: Predicting exam scores based on study hours
study_hours = np.array([1, 2, 3, 4, 5])
exam_scores = np.array([52, 57, 61, 67, 73])

# Calculate the slope (m) and intercept (b)
m, b = np.polyfit(study_hours, exam_scores, 1)

# Predicting values
predicted_scores = m * study_hours + b

# Printing thw equation
print(f"Regression line: y = {m:.2f}x + {b:.2f}")
```

```python
# Plotting the points and the regression line
plt.scatter(study_hours, exam_scores, label='Actual Data', color='blue')
plt.plot(study_hours, predicted_scores, label='Regression Line', color='red')
plt.xlabel('Study Hours')
plt.ylabel('Exam Scores')
plt.title('Linear Regression: Study Hours vs Exam Scores')
plt.legend()
plt.grid(True)
plt.show()
```

##  Explanation:

In this plot:
- The **blue dots** are the real data points.
- The **red line** is the predicted relationship using linear regression.

This shows us that the more you study, the higher your exam score tends to be — and regression helps us **quantify** that relationship!

 We can now **predict** exam scores for other study hours using this model. For example, if someone studies 6 hours, we just plug it into the formula!

##  Hypothesis Testing

Hypothesis testing helps us **make decisions** or **draw conclusions** about a population based on a sample.

We assume something is true (the **null hypothesis**), then use data to test that assumption.

---

##  Key Terms

- **Null Hypothesis (H₀)**: There’s no effect or difference (status quo).
- **Alternative Hypothesis (H₁)**: There **is** an effect or difference.
- **p-value**: The probability of observing our results (or more extreme) if H₀ were true.
- **Significance level (α)**: The threshold for rejecting H₀, often **0.05**.

---

##  Decision Rule

- If **p-value < 0.05** → Reject the null hypothesis (Result is significant)
- If **p-value ≥ 0.05** → Fail to reject the null hypothesis (Not enough evidence)

---

##  Errors in Testing

$
\begin{align*}
\text{Type I Error} & : \text{Rejecting } H_0 \text{ when it is actually true} \\
\text{Type II Error} & : \text{Failing to reject } H_0 \text{ when } H_1 \text{ is actually true}
\end{align*}
$

*Again.. Don’t worry, the code handles all the formulas!*


```python
import numpy as np
# We use scipy
from scipy import stats

# Old method scores (sample 1)
old_scores = np.array([67, 70, 72, 66, 69])

# New method scores (sample 2)
new_scores = np.array([75, 78, 74, 77, 76])

# Perform an independent two-sample t-test
t_stat, p_value = stats.ttest_ind(new_scores, old_scores)

print(f"T-Statistic: {t_stat:.2f}")
print(f"P-Value: {p_value:.4f}")
```

##  Explanation:

We used a **t-test** to compare the average scores of two teaching methods.

- The **T-Statistic** tells us how different the groups are.
- The **P-Value** tells us whether that difference is statistically significant.

In this example:
- If the p-value is less than 0.05 → The new method is significantly better!
- If not → The difference might be due to chance.

This is how **scientific research, A/B testing**, and even **medical trials** work.

##  ANOVA (Analysis of Variance)

When we want to **compare more than two group means**, ANOVA is the way to go.

Imagine you're testing **3 diets** and want to know if any of them make a difference in weight loss. Instead of doing many t-tests, we use ANOVA.

---

###  Hypotheses:
- **H₀ (null hypothesis)**: All group means are equal.
- **H₁ (alternative hypothesis)**: At least one group mean is different.

---

###  Formula (just for intuition)


$F = \frac{\text{Between-group variability}}{\text{Within-group variability}}$


 *Don't worry!!!!!!!!, the code does this for you.*

If the **F-statistic is large** and the **p-value is small**, it means there are significant differences between the groups.

##  Chi-Square Test

The **Chi-Square (χ²) Test** is used when you want to check if **two categorical variables are related** or independent.

---

###  Example use cases:
- Is there a relationship between **gender** and **preference for a product**?
- Is **city** related to **internet usage habits**?

---

###  Hypotheses:
- **H₀ (null hypothesis)**: The two variables are independent (no relationship)
- **H₁ (alternative hypothesis)**: The two variables are related

---

###  Chi-Square Formula:

$
\chi^2 = \sum \frac{(O - E)^2}{E}
$

Where:
- **O** = Observed frequency
- **E** = Expected frequency (what you'd expect if variables were independent)

*Maybe it looks hard but we’ll do it easily with `scipy` below.*


```python
import numpy as np
from scipy.stats import chi2_contingency

# A contingency table: Gender vs Product Preference
# Rows: Male, Female
# Columns: Prefer A, Prefer B
data = np.array([[30, 10],   # Males
                 [20, 40]])  # Females

# Perform Chi-Square test
chi2, p, dof, expected = chi2_contingency(data)

print(f"Chi2 Statistic: {chi2:.2f}")
print(f"P-Value: {p:.4f}")
print("Expected frequencies:")
print(expected)
```

##  Explanation:

We had a small study where males and females chose between two products (A or B).

- The **Chi2 statistic** tells us how far the observed data is from the expected values.
- The **P-value** tells us if the difference is significant.

---

###  Interpretation:
- If **p < 0.05**, we reject the null hypothesis → there is a relationship between gender and product choice.
- If **p ≥ 0.05**, we fail to reject the null → the choice is independent of gender.

---

###  Where it's used:
- Market surveys (brand vs location)
- Social science (education level vs opinion)
- Medical studies (treatment group vs result)
